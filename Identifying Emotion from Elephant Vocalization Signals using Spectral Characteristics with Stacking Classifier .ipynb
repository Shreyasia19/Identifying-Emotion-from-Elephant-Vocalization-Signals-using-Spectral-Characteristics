```python
# =================== Libraries ===================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier, ExtraTreesClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import RandomOverSampler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.multiclass import OneVsRestClassifier

# =================== Load Data ===================
file_path = "/content/20240106_dfall_obs_data_and_cepstral_features_revision1_n469.xlsx"
df = pd.read_excel(file_path)

# =================== Handle Missing Data ===================
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

# =================== Feature-Target Split ===================
X = df.drop(columns=['Context'])
y = df['Context']

# Encode target labels
le = LabelEncoder()
y = le.fit_transform(y)

# Optional: Class Distribution Plot
plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title("Target Class Distribution (Encoded)")
plt.show()

# =================== Data Preprocessing ===================
X = pd.get_dummies(X)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

ros = RandomOverSampler(random_state=42)
X_bal, y_bal = ros.fit_resample(X_scaled, y)

# =================== 70:15:15 Split ===================
X_train_val, X_test, y_train_val, y_test = train_test_split(X_bal, y_bal, test_size=0.15, stratify=y_bal, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1765, stratify=y_train_val, random_state=42)

print(f"Train shape: {X_train.shape}")
print(f"Validation shape: {X_val.shape}")
print(f"Test shape: {X_test.shape}")

# Check number of classes
n_classes = len(np.unique(y))
if n_classes == 2:
    y_test_bin = label_binarize(y_test, classes=[0, 1])
else:
    y_test_bin = label_binarize(y_test, classes=np.arange(n_classes))

# =================== Models and Parameters ===================
models = {
    'Neural Network': (MLPClassifier(max_iter=1000), {
        'hidden_layer_sizes': [(50,), (100,)],
        'learning_rate_init': [0.001, 0.01]
    }),
    'Logistic Regression': (LogisticRegression(max_iter=1000), {
        'C': [0.1, 1, 10]
    }),
    'Naive Bayes': (GaussianNB(), {}),
    'AdaBoost': (AdaBoostClassifier(), {
        'n_estimators': [50, 100],
        'learning_rate': [0.5, 1.0]
    }),
    'Bagging': (BaggingClassifier(), {
        'n_estimators': [50, 100]
    }),
    'Extra Trees': (ExtraTreesClassifier(), {
        'n_estimators': [50, 100],
        'max_depth': [5, 10]
    })
}

train_accuracies = {}
test_accuracies = {}

# =================== Model Training and Evaluation ===================
for name, (model, param_grid) in models.items():
    print(f"\n==================== {name} ====================")

    grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_model = grid.best_estimator_

    y_val_pred = best_model.predict(X_val)
    y_test_pred = best_model.predict(X_test)

    train_acc = accuracy_score(y_val, y_val_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    train_accuracies[name] = train_acc
    test_accuracies[name] = test_acc

    print(f"Best Parameters: {grid.best_params_}")
    print(f"Validation Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    print("Classification Report (Test Data):")
    print(classification_report(y_test, y_test_pred))

    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix (Test Data)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    if hasattr(best_model, "predict_proba"):
        y_score = best_model.predict_proba(X_test)
    else:
        y_score = OneVsRestClassifier(best_model).fit(X_train, y_train).predict_proba(X_test)

    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    plt.figure()

    if n_classes == 2:
        fpr[0], tpr[0], _ = roc_curve(y_test, y_score[:, 1])
        roc_auc[0] = auc(fpr[0], tpr[0])
        plt.plot(fpr[0], tpr[0], label=f'Class 1 (AUC = {roc_auc[0]:.2f})')
    else:
        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
            plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"{name} - ROC Curve (Test Data)")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

# =================== Bar Plot of Accuracies ===================
plt.figure(figsize=(10, 6))
x = np.arange(len(models))
width = 0.35

plt.bar(x - width/2, [train_accuracies[m] for m in models.keys()], width, label='Validation Accuracy')
plt.bar(x + width/2, [test_accuracies[m] for m in models.keys()], width, label='Test Accuracy')

plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Validation and Test Accuracies')
plt.xticks(x, models.keys(), rotation=45)
plt.ylim(0, 1.0)
plt.legend()
for i, v in enumerate([train_accuracies[m] for m in models.keys()]):
    plt.text(i - 0.2, v + 0.01, f"{v:.2f}", color='blue')
for i, v in enumerate([test_accuracies[m] for m in models.keys()]):
    plt.text(i + 0.05, v + 0.01, f"{v:.2f}", color='red')
plt.tight_layout()
plt.show()
```

```python
# =================== Libraries ===================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from imblearn.over_sampling import RandomOverSampler
from sklearn.multiclass import OneVsRestClassifier
from catboost import CatBoostClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.utils import to_categorical
import tensorflow as tf

# =================== Load Data ===================
file_path = "/content/20240106_dfall_obs_data_and_cepstral_features_revision1_n469.xlsx"
df = pd.read_excel(file_path)

# =================== Handle Missing Data ===================
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)

# =================== Feature-Target Split ===================
X = df.drop(columns=['Context'])
y = df['Context']

le = LabelEncoder()
y = le.fit_transform(y)

plt.figure(figsize=(6, 4))
sns.countplot(x=y)
plt.title("Target Class Distribution (Encoded)")
plt.show()

X = pd.get_dummies(X)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

ros = RandomOverSampler(random_state=42)
X_bal, y_bal = ros.fit_resample(X_scaled, y)

X_train_val, X_test, y_train_val, y_test = train_test_split(X_bal, y_bal, test_size=0.15, stratify=y_bal, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1765, stratify=y_train_val, random_state=42)

n_classes = len(np.unique(y))
y_test_bin = label_binarize(y_test, classes=np.arange(n_classes))
y_train_cat = to_categorical(y_train, num_classes=n_classes)
y_val_cat = to_categorical(y_val, num_classes=n_classes)
y_test_cat = to_categorical(y_test, num_classes=n_classes)

# =================== Models ===================
models = {
    'CatBoost': CatBoostClassifier(verbose=0),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, learning_rate=0.05),
    'MLP': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)
}

train_accuracies = {}
test_accuracies = {}

# =================== Training and Evaluation ===================
for name, model in models.items():
    print(f"\n==================== {name} ====================")

    model.fit(X_train, y_train)
    y_val_pred = model.predict(X_val)
    y_test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_val, y_val_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    train_accuracies[name] = train_acc
    test_accuracies[name] = test_acc

    print(f"Validation Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(classification_report(y_test, y_test_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # ROC Curve
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)
    else:
        y_score = OneVsRestClassifier(model).fit(X_train, y_train).predict_proba(X_test)

    fpr, tpr, roc_auc = {}, {}, {}
    plt.figure()
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(f"{name} - ROC Curve")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

# =================== DNN ===================
print("\n==================== DNN ====================")
model_dnn = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(n_classes, activation='softmax')
])
model_dnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history_dnn = model_dnn.fit(X_train, y_train_cat, epochs=50, batch_size=16, verbose=0, validation_data=(X_val, y_val_cat))

train_acc_dnn = history_dnn.history['val_accuracy'][-1]
test_acc_dnn = model_dnn.evaluate(X_test, y_test_cat, verbose=0)[1]
train_accuracies['DNN'] = train_acc_dnn
test_accuracies['DNN'] = test_acc_dnn

print(f"Validation Accuracy: {train_acc_dnn:.4f}")
print(f"Test Accuracy: {test_acc_dnn:.4f}")

# Confusion Matrix DNN
y_test_pred_dnn = np.argmax(model_dnn.predict(X_test), axis=1)
cm = confusion_matrix(y_test, y_test_pred_dnn)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("DNN - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve DNN
fpr, tpr, roc_auc = {}, {}, {}
y_score_dnn = model_dnn.predict(X_test)
plt.figure()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_dnn[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title("DNN - ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(True)
plt.show()

# =================== LSTM ===================
print("\n==================== LSTM ====================")
X_train_lstm = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val_lstm = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))
X_test_lstm = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

model_lstm = Sequential([
    LSTM(64, input_shape=(X_train_lstm.shape[1], 1)),
    Dense(n_classes, activation='softmax')
])
model_lstm.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
history_lstm = model_lstm.fit(X_train_lstm, y_train_cat, epochs=50, batch_size=16, verbose=0, validation_data=(X_val_lstm, y_val_cat))

train_acc_lstm = history_lstm.history['val_accuracy'][-1]
test_acc_lstm = model_lstm.evaluate(X_test_lstm, y_test_cat, verbose=0)[1]
train_accuracies['LSTM'] = train_acc_lstm
test_accuracies['LSTM'] = test_acc_lstm

print(f"Validation Accuracy: {train_acc_lstm:.4f}")
print(f"Test Accuracy: {test_acc_lstm:.4f}")

# Confusion Matrix LSTM
y_test_pred_lstm = np.argmax(model_lstm.predict(X_test_lstm), axis=1)
cm = confusion_matrix(y_test, y_test_pred_lstm)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("LSTM - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# ROC Curve LSTM
fpr, tpr, roc_auc = {}, {}, {}
y_score_lstm = model_lstm.predict(X_test_lstm)
plt.figure()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_lstm[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.title("LSTM - ROC Curve")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(True)
plt.show()

# =================== Final Accuracy Bar Plot ===================
plt.figure(figsize=(10, 6))
x = np.arange(len(train_accuracies))
width = 0.35

plt.bar(x - width/2, list(train_accuracies.values()), width, label='Validation Accuracy')
plt.bar(x + width/2, list(test_accuracies.values()), width, label='Test Accuracy')

plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Validation and Test Accuracies')
plt.xticks(x, train_accuracies.keys(), rotation=45)
plt.ylim(0, 1.0)
plt.legend()

for i, v in enumerate(train_accuracies.values()):
    plt.text(i - 0.2, v + 0.01, f"{v:.2f}", color='blue')
for i, v in enumerate(test_accuracies.values()):
    plt.text(i + 0.05, v + 0.01, f"{v:.2f}", color='red')

plt.tight_layout()
plt.show()
```

AutoEncoder

“We employed a self-supervised autoencoder to learn compressed representations of the vocalization spectral features. The latent embeddings (bottleneck layer) capture high-level acoustic patterns, enabling downstream classifiers to generalize better. This representation learning approach enhances both performance and interpretability.”

```python


import random
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay

# Load dataset
df = pd.read_excel('/content/20240106_dfall_obs_data_and_cepstral_features_revision1_n469.xlsx')

# 1. Select only spectral features
spectral_features = ['sprsMed', 'sprsMbw', 'sprsEqbw', 'sprsMc'] + \
                    [f'V{i}' for i in range(1, 13)] + \
                    [f'Fw{i}' for i in range(1, 5)] + \
                    [f'Mw{i}' for i in range(1, 5)] + \
                    [f'F{i}' for i in range(1, 5)] + \
                    [f'M{i}' for i in range(1, 5)]

X = df[spectral_features]
y = df['Context']

# 1. Set random seeds
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# 2. Drop rows where features or labels are missing
data = pd.concat([X, y], axis=1)
data = data.dropna()

X = data[spectral_features]
y = data['Context']

# 3. Remove rare classes (classes with < 2 samples)
valid_classes = y.value_counts()[y.value_counts() >= 2].index
X = X[y.isin(valid_classes)]
y = y[y.isin(valid_classes)]

# 4. Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# 6. Build Autoencoder
input_dim = X_train.shape[1]
input_layer = Input(shape=(input_dim,))
# Try bottleneck = 32 instead of 16
encoded = Dense(64, activation='relu')(input_layer)
bottleneck = Dense(32, activation='relu')(encoded)

decoded = Dense(32, activation='relu')(bottleneck)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# 7. Train Autoencoder
history = autoencoder.fit(X_train, X_train,
                          epochs=100,
                          batch_size=16,
                          shuffle=True,
                          validation_data=(X_test, X_test),
                          verbose=1)

# 8. Extract bottleneck features
encoder = Model(inputs=input_layer, outputs=bottleneck)
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# 9. Train classifier on encoded features
clf = RandomForestClassifier(n_estimators=200, random_state=42)
clf.fit(X_train_encoded, y_train)

# Evaluate classifier
y_pred = clf.predict(X_test_encoded)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
labels = sorted(y.unique())  # Make sure labels match order

# Plot confusion matrix as heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted Context')
plt.ylabel('True Context')
plt.title('Confusion Matrix Heatmap')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Also print classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
